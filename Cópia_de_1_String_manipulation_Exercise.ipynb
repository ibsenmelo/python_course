Gerando uma função que vai contar as palavras:

```python
def word_count(str):
    counts = dict()
    words = str.split()

    for word in words:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1

    return counts
```

Retirando o que não é letra para não atrapalhar na identificação de palavras únicas:

```python
gene = re.sub("[^a-zA-Z]+", " ", genesis)
```

```python
aa1 = word_count(gene)
```

```python
a = {k: v for k, v in sorted(aa1.items(), key=lambda x: x[1], reverse=True)}
```

```python
n = 5
{k: a[k] for k in list(a)[:n]}
```
